```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#df = pd.read_excel('cmip6_djf_79_14.xlsx')  DJF
df = pd.read_excel('ond_cmpi6_1901_2014.xlsx')  #OND
#sampl_data
samples = df.year[:113]
x=np.arange(len(samples))
obs = df.IMD[:113]
mme = df.MME[:113]
#print(samples)
```

```python
bar_width = 0.5
plt.figure(figsize=(18,4),constrained_layout=False) 
#plt.bar(samples,obs)
opacity = 0.8
rects1 = plt.bar(samples,obs, bar_width,alpha=opacity,color='black',label='OBS')
rects2 = plt.bar(samples + bar_width, mme, bar_width,alpha=opacity,color='gray',label='MME')

plt.axhline(y=0.5, color='black', linestyle='--')
plt.axhline(y=-0.5, color='black', linestyle='--')
plt.xlabel('Years',fontsize=12,fontweight='bold')
plt.ylabel('Normalized Anomalies',fontsize=12,fontweight='bold')
#plt.title('Inter-annual(DJF)_variability',fontsize=12,fontweight='bold')
plt.tick_params(labelsize = 15,width=3)
plt.legend()
plt.savefig('ond_interannula_variability.jpg', bbox_inches='tight', dpi=300)
plt.tight_layout()
plt.show()

```
